```python
import pandas as pd
import numpy as np

space_size = 10

factory_layout = pd.DataFrame(np.full((space_size, space_size), ' '),
                              columns=[f'C{i+1}' for i in range(space_size)],
                              index=[f'R{i+1}' for i in range(space_size)])

A_machines = 10
B_machines = 40
C_machines = 15

a_machine_id = 1
row, col = space_size - 1, space_size - 1

while A_machines > 0:
    if col - 1 >= 0 and (row != space_size // 2 and col != space_size // 2):
        factory_layout.iloc[row, col] = f'A{a_machine_id}'
        factory_layout.iloc[row, col-1] = f'A{a_machine_id}'
        col -= 2
    else:
        row -= 1
        if row == space_size // 2:
            row -= 1
        col = space_size - 1
        if col - 1 >= 0:
            factory_layout.iloc[row, col] = f'A{a_machine_id}'
            factory_layout.iloc[row, col-1] = f'A{a_machine_id}'
            col -= 2
    a_machine_id += 1
    A_machines -= 1


b_machine_id = 1

for r in range(space_size):
    for c in range(space_size):
        if factory_layout.iloc[r, c] == ' ' and (r != space_size // 2 and c != space_size // 2) and B_machines > 0:
            factory_layout.iloc[r, c] = f'B{b_machine_id}'
            b_machine_id += 1
            B_machines -= 1

c_machine_id = 1

for r in range(space_size):
    for c in range(space_size):
        if factory_layout.iloc[r, c] == ' ' and (r != space_size // 2 and c != space_size // 2) and C_machines > 0:
            factory_layout.iloc[r, c] = f'C{c_machine_id}'
            c_machine_id += 1
            C_machines -= 1


print("Factory Layout:")
print(factory_layout)
```

```python
import pandas as pd
import numpy as np

# 定義工廠空間大小
space_rows = 10
space_cols = 10

# 創建空的工廠佈局
factory_layout = pd.DataFrame(np.full((space_rows, space_cols), ' '),
                              columns=[f'C{i+1}' for i in range(space_cols)],
                              index=[f'R{i+1}' for i in range(space_rows)])

# 定義機台信息 (名稱, 佔地面積, 重量)
machines = [
    ('A1', 2, 10), ('A2', 2, 10), ('A3', 2, 10),
    ('B1', 1, 5), ('B2', 1, 5), ('B3', 1, 5), ('B4', 1, 5),
    ('C1', 1, 3), ('C2', 1, 3), ('C3', 1, 3), ('C4', 1, 3),
    ('C5', 1, 3), ('C6', 1, 3), ('C7', 1, 3), ('C8', 1, 3),
    ('C9', 1, 3), ('C10', 1, 3), ('C11', 1, 3), ('C12', 1, 3),
    ('C13', 1, 3), ('C14', 1, 3), ('C15', 1, 3)
]

# 定義樓層承載重量
max_weight_per_floor = 100

# 定義支撐柱子的位置 (行, 列)
columns = [(2, 2), (5, 5), (7, 7)]


for col in columns:
    factory_layout.iloc[col[0], col[1]] = 'P'

def calculate_current_weight(factory_layout):
    current_weight = 0
    for row in range(factory_layout.shape[0]):
        for col in range(factory_layout.shape[1]):
            cell = factory_layout.iloc[row, col]
            if cell.startswith('A'):
                current_weight += 10
            elif cell.startswith('B'):
                current_weight += 5
            elif cell.startswith('C'):
                current_weight += 3
    return current_weight


def place_machine(factory_layout, machine, weight_limit):
    name, size, weight = machine
    for row in range(space_rows):
        for col in range(space_cols):
            if can_place(factory_layout, row, col, size):
                if calculate_current_weight(factory_layout) + weight <= weight_limit:
                    factory_layout = place(factory_layout, row, col, name, size)
                    return True, factory_layout
    return False, factory_layout


def can_place(factory_layout, row, col, size):
    if col + size - 1 >= space_cols:
        return False
    for i in range(size):
        if factory_layout.iloc[row, col + i] != ' ':
            return False
    return True


def place(factory_layout, row, col, name, size):
    for i in range(size):
        factory_layout.iloc[row, col + i] = name
    return factory_layout


for machine in machines:
    placed, factory_layout = place_machine(factory_layout, machine, max_weight_per_floor)
    if not placed:
        print(f"Unable to place machine {machine[0]} due to space or weight constraints.")

print("Factory Layout:")
print(factory_layout)
```
